## 4. ğŸ“Š Load Data

Let's use the famous **Iris** dataset for our first example.

### About the Iris Dataset

- ğŸŒ¸ **150 samples** of Iris flowers
- ğŸ¯ **3 classes**: Setosa, Versicolor, Virginica
- ğŸ“ **4 features**: Sepal Length, Sepal Width, Petal Length, Petal Width
- âœ… **Clean and balanced** - perfect for learning!

Let's load and visualize:

````python
# Load Iris dataset
iris = load_iris()

# Create DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Map classes to readable names
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print("ğŸ“Š Iris dataset loaded!")
print(f"\nğŸ“ Shape: {df.shape}")
print(f"ğŸ“‹ Columns: {df.columns.tolist()}")
print(f"\nğŸ¯ Classes:")
print(df['species'].value_counts())
````

### Descriptive statistics

## 5. ğŸ Create DBDataset

Now comes the interesting part! Let's create our first **DBDataset**.

### What is DBDataset?

`DBDataset` is the **fundamental container** of DeepBridge that encapsulates:
- ğŸ“¦ Training and test data
- ğŸ¯ Features and target
- ğŸ¤– Model (optional)
- ğŸ“Š Predictions (if model provided)
- ğŸ·ï¸ Metadata (categorical features, etc.)

### Create DBDataset

It's very simple! Just 3 required parameters:

````python
# Create DBDataset
dataset = DBDataset(
    data=df,                    # Complete DataFrame
    target_column='target',     # Target column
    test_size=0.2,              # 20% for testing
    random_state=42             # For reproducibility
)

print("âœ… DBDataset created successfully!")
print(f"\nğŸ“Š Details:")
print(f"   Training samples: {len(dataset.train_data)}")
print(f"   Test samples: {len(dataset.test_data)}")
print(f"   Total: {len(dataset.train_data) + len(dataset.test_data)}")
````

### What happened?

DBDataset automatically:
1. âœ… Split the data into train (80%) and test (20%)
2. âœ… Identified features (everything except 'target')
3. âœ… Detected categorical features
4. âœ… Validated data consistency

**All of this with one line of code!** ğŸ‰

<a id="explore"></a>
## 6. ğŸ” Explore DBDataset

Let's explore the properties of our DBDataset:

### 6.1 Access Training and Test Data

````python
# Training data
print("ğŸ“¦ Training Data:")
print(f"   Shape: {dataset.train_data.shape}")
print(f"\n   First rows:")
display(dataset.train_data.head())
````

````python
# Test data
print("ğŸ“¦ Test Data:")
print(f"   Shape: {dataset.test_data.shape}")
print(f"\n   First rows:")
display(dataset.test_data.head())
````

### 6.2 Features and Target

````python
# Feature list
print("ğŸ“‹ Features:")
for i, feat in enumerate(dataset.features, 1):
    print(f"   {i}. {feat}")

print(f"\nğŸ¯ Target: {dataset.target_name}")
print(f"\nğŸ“Š Total features: {len(dataset.features)}")
````

### 6.3 Categorical vs Numerical Features

````python
# Categorical features (auto-detected)
print("ğŸ·ï¸ Categorical Features:")
if dataset.categorical_features:
    for feat in dataset.categorical_features:
        print(f"   - {feat}")
else:
    print("   No categorical features detected")

# Numerical features
print("\nğŸ“Š Numerical Features:")
for feat in dataset.numerical_features:
    print(f"   - {feat}")
````

<div style="background-color: #d4edda; padding: 10px; border-radius: 5px; border-left: 5px solid #28a745;">
<b>âœ… Amazing!</b> DBDataset automatically detected that all features are numerical!
</div>

### 6.4 Access Features and Target Separately

````python
# Get training features (X_train)
X_train = dataset.get_feature_data('train')
print("ğŸ“Š X_train (training features):")
print(f"   Shape: {X_train.shape}")
print(f"   Columns: {X_train.columns.tolist()}")

# Get training target (y_train)
y_train = dataset.get_target_data('train')
print(f"\nğŸ¯ y_train (training target):")
print(f"   Shape: {y_train.shape}")
print(f"   Unique values: {y_train.unique()}")
````

````python
# Similar for test
X_test = dataset.get_feature_data('test')
y_test = dataset.get_target_data('test')

print("ğŸ“Š Test Set:")
print(f"   X_test shape: {X_test.shape}")
print(f"   y_test shape: {y_test.shape}")
````

<a id="viz"></a>
## 7. ğŸ“ˆ Visualizations

Let's create some visualizations to better understand our dataset:

### 7.1 Class Distribution

````python
# Count classes in train and test
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Train
y_train.value_counts().plot(kind='bar', ax=axes[0], color='skyblue', edgecolor='black')
axes[0].set_title('Class Distribution - Train', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Class')
axes[0].set_ylabel('Count')
axes[0].grid(axis='y', alpha=0.3)

# Test
y_test.value_counts().plot(kind='bar', ax=axes[1], color='lightcoral', edgecolor='black')
axes[1].set_title('Class Distribution - Test', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Class')
axes[1].set_ylabel('Count')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("âœ… Balanced dataset - great for training!")
````

### 7.2 Feature Distribution

````python
# Feature histograms
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()

for i, feature in enumerate(dataset.features[:4]):  # First 4 features
    X_train[feature].hist(ax=axes[i], bins=20, color='steelblue', edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Distribution: {feature}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')
    axes[i].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
````

### 7.3 Feature Correlation

````python
# Correlation matrix
plt.figure(figsize=(10, 8))
correlation = X_train.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix - Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nğŸ’¡ Observation: Highly correlated features may contain redundant information!")
````
