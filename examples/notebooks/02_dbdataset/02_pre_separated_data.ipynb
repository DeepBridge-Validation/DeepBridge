## ğŸ¯ Objectives
- âœ… Use pre-separated datasets
- âœ… DBDataset with train_data + test_data
- âœ… Automatic validations

```python
import pandas as pd
import numpy as np
from deepbridge import DBDataset
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Simulate pre-separated data
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target

# Manual split (as if coming from different files)
train, test = train_test_split(df, test_size=0.2, random_state=42)

print(f"Train: {train.shape}")
print(f"Test: {test.shape}")
```

## Create DBDataset with Pre-separated Data

```python
# METHOD 2: With train_data and test_data
dataset = DBDataset(
    train_data=train,  # Training DataFrame
    test_data=test,    # Test DataFrame
    target_column='target'
)

print("âœ… DBDataset created with pre-separated data!")
print(f"\nTrain: {len(dataset.train_data)} samples")
print(f"Test: {len(dataset.test_data)} samples")
```

## Automatic Validations

```python
# DBDataset validates automatically:
print("ğŸ” Automatic Validations:\n")
print("âœ… 1. Train and test columns are consistent")
print("âœ… 2. Target exists in both datasets")
print("âœ… 3. Features are the same")
print("âœ… 4. Compatible data types")

print(f"\nğŸ“Š Features detected: {len(dataset.features)}")
print(f"ğŸ“‹ Features: {dataset.features[:5]}...")
```

## Compare Both Methods

```python
# METHOD 1: Automatic split
ds1 = DBDataset(data=df, target_column='target', test_size=0.2, random_state=42)

# METHOD 2: Pre-separated
ds2 = DBDataset(train_data=train, test_data=test, target_column='target')

print("Method 1 (automatic split):")
print(f"  Train: {len(ds1.train_data)}, Test: {len(ds1.test_data)}")

print("\nMethod 2 (pre-separated):")
print(f"  Train: {len(ds2.train_data)}, Test: {len(ds2.test_data)}")

print("\nğŸ’¡ Use Method 1 when you have a single DataFrame")
print("ğŸ’¡ Use Method 2 when data is already separated (e.g., Kaggle)")
```
